```python
import math
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load spreadsheet
data_path = r"c:\Users\Chris Yeung\Queen's University\Tamas Ungi - ChrisYTamasShare\2025_MarginsStudy\NaviKnife_SyncInfo.xlsx"
```

```python
iknife_df = pd.read_excel(data_path, sheet_name="iKnife")
print(len(iknife_df))
iknife_df.head()
```

```python
# replace "-" with NaN
iknife_df = iknife_df.replace('-', np.nan)
iknife_df.head()
```

```python
# remove columns after Notes
iknife_df = iknife_df.loc[:, :'Notes']
# remove specified columns
cols_to_rm = ["time", "CR#", "case # (lump. Consent / mast iKnife)", "Slicer scene?", "Position matrix?"]
iknife_df = iknife_df.drop(cols_to_rm, axis=1)
iknife_df.head()
```

```python
# remove rows without a margin status
iknife_df = iknife_df.dropna(subset=['Margin status (IDC)', 'Margin status (DCIS)'])
iknife_df = iknife_df.reset_index(drop=True)
print(len(iknife_df))
iknife_df.tail()
```

```python
# convert margin status to integer
iknife_df['Margin status (IDC)'] = iknife_df['Margin status (IDC)'].apply(lambda x: 1 if x == 'Pos' else 0)
iknife_df['Margin status (DCIS)'] = iknife_df['Margin status (DCIS)'].apply(lambda x: 1 if x == 'Pos' else 0)
iknife_df.head()
```

```python
def calculate_volume(dimensions):
    total_volume = 0
    if pd.isnull(dimensions):
        return
    
    si, ml, ap = map(float, dimensions.split('x'))
    volume = (4/3) * math.pi * (si/2) * (ml/2) * (ap/2)
    total_volume += volume
    return total_volume
```

```python
iknife_df['Specimen volume (cm3)'] = iknife_df['Specimen dimensions (cm)'].apply(calculate_volume)
iknife_df['Tumor volume (cm3)'] = iknife_df['Tumor dimensions (cm)'].apply(calculate_volume)
iknife_df['Normalized specimen volume (cm3)'] = iknife_df['Specimen volume (cm3)'] / iknife_df['Tumor volume (cm3)']
iknife_df.head()
```

```python
# read sheet containing previous nav cases
gauvin_df = pd.read_excel(data_path, sheet_name="Gauvin2019")
print(len(gauvin_df))
gauvin_df.head()
```

```python
# remove specified columns
cols_to_rm = ["time", "CR#", "case # (lump. Consent / mast iKnife)", "Slicer scene?", "Position matrix?"]
gauvin_df = gauvin_df.drop(cols_to_rm, axis=1)
gauvin_df.head()
```

```python
# remove rows without a margin status
gauvin_df = gauvin_df.dropna(subset=['Side'])
gauvin_df = gauvin_df.reset_index(drop=True)
print(len(gauvin_df))
gauvin_df.tail()
```

```python
# convert margin status columns to integer
gauvin_df['Margin status (IDC)'] = gauvin_df['Margin status (IDC)'].apply(lambda x: 1 if x == 'Pos' else 0)
gauvin_df['Margin status (DCIS)'] = gauvin_df['Margin status (DCIS)'].apply(lambda x: 1 if x == 'Pos' else 0)
gauvin_df.head()
```

```python
# calculate volume
gauvin_df['Specimen volume (cm3)'] = gauvin_df['Specimen dimensions (cm)'].apply(calculate_volume)
gauvin_df['Tumor volume (cm3)'] = gauvin_df['Tumor dimensions (cm)'].apply(calculate_volume)
gauvin_df['Normalized specimen volume (cm3)'] = gauvin_df['Specimen volume (cm3)'] / gauvin_df['Tumor volume (cm3)']
gauvin_df.head()
```

```python
total_df = pd.concat([iknife_df, gauvin_df], ignore_index=True)
print(len(total_df))
total_df.head()
```

```python
# count how percentage of cases have positive margins (from + (IDC or DCIS) column)
total_df['+ (IDC or DCIS)'] = total_df['+ (IDC or DCIS)'].astype(int)
total_pos_rate = total_df['+ (IDC or DCIS)'].value_counts(normalize=True).get(1, 0)
print(f"Total positive margin rate: {total_pos_rate:.2%}")
# print the count out of the total number of cases
total_df['+ (IDC or DCIS)'].value_counts()
```

```python
from scipy.stats import fisher_exact

# calculate positive margin rate separately for iknife and gauvin cases
iknife_pos_rate = iknife_df['+ (IDC or DCIS)'].value_counts(normalize=True).get(1, 0)
gauvin_pos_rate = gauvin_df['+ (IDC or DCIS)'].value_counts(normalize=True).get(1, 0)
print(f"iKnife positive margin rate: {iknife_pos_rate:.2%}")
print(f"Gauvin positive margin rate: {gauvin_pos_rate:.2%}")
# show count of positive margins for each group
print(iknife_df['+ (IDC or DCIS)'].value_counts())
print(gauvin_df['+ (IDC or DCIS)'].value_counts())
# create contingency table
contingency_table = [
    [gauvin_df['+ (IDC or DCIS)'].count() - gauvin_df['+ (IDC or DCIS)'].sum(), gauvin_df['+ (IDC or DCIS)'].sum()],
    [iknife_df['+ (IDC or DCIS)'].count() - iknife_df['+ (IDC or DCIS)'].sum(), iknife_df['+ (IDC or DCIS)'].sum()]
]
print(contingency_table)

# perform Fisher's exact test
odds_ratio, p_value = fisher_exact(contingency_table, alternative="less")
print(f"Odds ratio: {odds_ratio:.4f}")
print(f"Fisher's exact test p-value: {p_value:.4f}")
```

```python
# calculate mean specimen volume
total_mean_volume = total_df['Specimen volume (cm3)'].mean()
iknife_mean_volume = iknife_df['Specimen volume (cm3)'].mean()
gauvin_mean_volume = gauvin_df['Specimen volume (cm3)'].mean()
print(f"Total mean specimen volume: {total_mean_volume:.2f}")
print(f"iKnife mean specimen volume: {iknife_mean_volume:.2f}")
print(f"Gauvin mean specimen volume: {gauvin_mean_volume:.2f}")
```

```python
# calculate mean normalized specimen volume
total_mean_norm_volume = total_df['Normalized specimen volume (cm3)'].mean()
iknife_mean_norm_volume = iknife_df['Normalized specimen volume (cm3)'].mean()
gauvin_mean_norm_volume = gauvin_df['Normalized specimen volume (cm3)'].mean()
print(f"Total mean normalized specimen volume: {total_mean_norm_volume:.2f}")
print(f"iKnife mean normalized specimen volume: {iknife_mean_norm_volume:.2f}")
print(f"Gauvin mean normalized specimen volume: {gauvin_mean_norm_volume:.2f}")
```

```python
# calculate mean specimen weight
total_mean_weight = total_df['Weight (g)'].mean()
iknife_mean_weight = iknife_df['Weight (g)'].mean()
gauvin_mean_weight = gauvin_df['Weight (g)'].mean()
print(f"Total mean specimen weight: {total_mean_weight:.2f}")
print(f"iKnife mean specimen weight: {iknife_mean_weight:.2f}")
print(f"Gauvin mean specimen weight: {gauvin_mean_weight:.2f}")
```
